# 예보데이터

```python
# 20시(세계시간11시)에 발표된 3시간 단위의 예보데이터

def make_pre_data(name,col_name):
    n = pd.read_csv('./청주시_흥덕구_복대1동_예보(2019.07-2020.06)/'+name+'.csv')
    n.columns = ['day','hour','forecast','value']
    n = n[(n.hour==1100.0) & (n.forecast>=4.0) & (n.forecast<=25.0)]
    n.reset_index(drop=True,inplace=True)

    t=[]
    length = len(n)
    for i in range(length):
        t.append(n.value[i])
        if i==length:
            break
        t.append(-1)
        t.append(-1)

    n = pd.DataFrame({col_name:t[:8782]})
    n[n[col_name]==-1] = np.nan
    if name == '하늘상태':
        n = n.fillna(method='ffill')
    else :
        n = round(n.interpolate(),1)
    return n
```

```python
df1 = make_pre_data('3시간기온','pre_temperature')
df2 = make_pre_data('풍속','pre_wind_speed')
df3 = make_pre_data('풍향','pre_wind_direction')
df4 = make_pre_data('습도','pre_humidity')
df5 = make_pre_data('하늘상태','pre_cloud')
df6 = make_pre_data('강수확률','pre_rain')

pre_weather = pd.concat([df1,df2,df3,df4,df5,df6],axis=1)
pre_weather
```

```python
def make_pre_data_0731(n,col_name):
    t=[]
    length = len(n)
    for i in range(length):
        t.append(n[col_name][i])
        if i==length-1:
            break
        t.append(-1)
        t.append(-1)

    n = pd.DataFrame({col_name:t})
    n[n[col_name]==-1] = np.nan
    if col_name == 'pre_cloud':
        n = n.fillna(method='ffill')
    else :
        n = round(n.interpolate(),1)
    return n
```

# 파생변수

## 계절/날씨-시간   평균/사분위수

```python
# 계절/시간 변수 만들기

season=[];time=[]
for i in list(range(len(data1))):
    cond = data1.index.month[i]
    if cond<=2 or cond==12:
        season.append('winter')
    elif cond<=5:
        season.append('spring')
    elif cond<=8:
        season.append('summer')
    elif cond<=11:
        season.append('fall')
    time.append(data1.index.hour[i])

season = pd.DataFrame({'season':season})
time = pd.DataFrame({'time':time})
```

```python
# 계절/시간 변수 합치기

data2 = data1.reset_index()
data2['date'] = pd.to_datetime(data2['date'])
data2 = pd.concat([data2,season,time],axis=1)
data2
```

```python
# 계절에 따른 동시간의 평균/사분위수

r_st_mean=[]
r_st_quantile1=[]
r_st_quantile2=[]
r_st_quantile3=[]
q_st_mean=[]
q_st_quantile1=[]
q_st_quantile2=[]
q_st_quantile3=[]


for i in list(range(len(data2))):
    s = data2['season'][i]
    t = data2['time'][i]
    rst = data2.loc[(data2['season']==s) & (data2['time']==t)].radiation
    qst = data2.loc[(data2['season']==s) & (data2['time']==t)].quantity
    
    rs = rst.mean()
    rs1 = rst.quantile(.25)
    rs2 = rst.quantile(.5)
    rs3 = rst.quantile(.75)
    
    qs = qst.mean()
    qs1 = qst.quantile(.25)
    qs2 = qst.quantile(.5)
    qs3 = qst.quantile(.75)
    
    r_st_mean.append(round(rs,3))
    r_st_quantile1.append(round(rs1,3))
    r_st_quantile2.append(round(rs2,3))
    r_st_quantile3.append(round(rs3,3))

    q_st_mean.append(round(qs,3))
    q_st_quantile1.append(round(qs1,3))
    q_st_quantile2.append(round(qs2,3))
    q_st_quantile3.append(round(qs3,3))
```

```python
dic1 = {'radiation_st_mean':r_st_mean,
       'radiation_st_quantile1':r_st_quantile1,
       'radiation_st_quantile2':r_st_quantile2,
       'radiation_st_quantile3':r_st_quantile3,
       'quantity_st_mean':q_st_mean,
       'quantity_st_quantile1':q_st_quantile1,
       'quantity_st_quantile2':q_st_quantile2,
       'quantity_st_quantile3':q_st_quantile3
      }
df_st = pd.DataFrame(dic1)
df_st
```

```python
# 하늘상태에 따른 동시간의 평균/사분위수

r_ct_mean=[]
r_ct_quantile1=[]
r_ct_quantile2=[]
r_ct_quantile3=[]
q_ct_mean=[]
q_ct_quantile1=[]
q_ct_quantile2=[]
q_ct_quantile3=[]

for i in list(range(len(data2))):
    c = data2['pre_cloud'][i]
    t = data2['time'][i]
    
    rct = data2.loc[(data2['pre_cloud']==c) & (data2['time']==t)].radiation
    qct = data2.loc[(data2['pre_cloud']==c) & (data2['time']==t)].quantity
    
    rc = rct.mean()
    rc1 = rct.quantile(.25)
    rc2 = rct.quantile(.5)
    rc3 = rct.quantile(.75)
    
    qc = qct.mean()
    qc1 = qct.quantile(.25)
    qc2 = qct.quantile(.5)
    qc3 = qct.quantile(.75)
   
    r_ct_mean.append(round(rc,3))
    r_ct_quantile1.append(round(rc1,3))
    r_ct_quantile2.append(round(rc2,3))
    r_ct_quantile3.append(round(rc3,3))
    
    q_ct_mean.append(round(qc,3))
    q_ct_quantile1.append(round(qc1,3))
    q_ct_quantile2.append(round(qc2,3))
    q_ct_quantile3.append(round(qc3,3))
```

```python
dic2 = {'radiation_ct_mean':r_ct_mean,
       'radiation_ct_quantile1':r_ct_quantile1,
       'radiation_ct_quantile2':r_ct_quantile2,
       'radiation_ct_quantile3':r_ct_quantile3,
       'quantity_ct_mean':q_ct_mean,
       'quantity_ct_quantile1':q_ct_quantile1,
       'quantity_ct_quantile2':q_ct_quantile2,
       'quantity_ct_quantile3':q_ct_quantile3
      }
df_ct = pd.DataFrame(dic2)
```

```python
data3 = pd.concat([data2,df_st,df_ct],axis=1)
del data3['season']
del data3['time']
data3
```

```python
# 결측치 확인

data3.isna().sum()
```

## 자기회귀

```python
# 7일간의 동시간 발전량 자기회귀

ar1 = data3.quantity.shift(24*1)
ar2 = data3.quantity.shift(24*2)
ar3 = data3.quantity.shift(24*3)
ar4 = data3.quantity.shift(24*4)
ar5 = data3.quantity.shift(24*5)
ar6 = data3.quantity.shift(24*6)
ar7 = data3.quantity.shift(24*7)

q_ar = pd.concat([ar1,ar2,ar3,ar4,ar5,ar6,ar7],axis=1)
q_ar.columns = ['q_ar1','q_ar2','q_ar3','q_ar4','q_ar5','q_ar6','q_ar7']
q_ar
```

```python
# 7일간의 동시간 일사량 자기회귀

ar1 = data3.radiation.shift(24*1)
ar2 = data3.radiation.shift(24*2)
ar3 = data3.radiation.shift(24*3)
ar4 = data3.radiation.shift(24*4)
ar5 = data3.radiation.shift(24*5)
ar6 = data3.radiation.shift(24*6)
ar7 = data3.radiation.shift(24*7)

r_ar = pd.concat([ar1,ar2,ar3,ar4,ar5,ar6,ar7],axis=1)
r_ar.columns = ['r_ar1','r_ar2','r_ar3','r_ar4','r_ar5','r_ar6','r_ar7']
r_ar.reset_index(drop=True,inplace=True)
r_ar
```

```python
ar = pd.concat([q_ar,r_ar],axis=1)
ar
```
